```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
import requests

```

```python
def scrape_info():

    pages = ["","p.2","p.3","p.4"]
    df = []
    
    for x in pages:
        url = 'https://steamcharts.com/top/'+x
        executable_path = {'executable_path': 'chromedriver.exe'}
        browser = Browser('chrome', **executable_path, headless=False)
        browser.visit(url)

    #     url = 'http://www.ffiec.gov/census/report.aspx?year=2011&state=01&report=demographic&msa=11500'
        html = requests.get(url).content
        df_list = pd.read_html(html)
        data = df_list[-1]
        df.append(data)
#         print(df)
        
        # len(states_values)
    #     for x in states_values:
    #         one_state = browser.links.find_by_href(f'https://geology.com/county-map/{x.lower()}.shtml').click()
    #         html = browser.html
    #         soup = bs(html,'html.parser')
    #         image = soup.find('img')['src']
    #         images_urls.append({"state":x, "image":image})
    #         print()
    #         time.sleep(5)
    #         browser.back()
        browser.quit()
    return df
df = scrape_info()
```

```python
def scrape_list(names_list):

#     pages = ["","p.2","p.3","p.4"]
    iteration = names_list
    
    df1=[]
    for i in iteration:
        url = 'https://steamcharts.com/top/'
        executable_path = {'executable_path': 'chromedriver.exe'}
        browser = Browser('chrome', **executable_path, headless=False)
        browser.visit(url)
        time.sleep(1)
    #     tablerow = browser.find_by_tag('a')
        print(f"we are looking for the name {i}")
        
        try:
            browser.find_link_by_partial_text(i).first.click()
        except:
            print(f"we have skiped {i}")
        time.sleep(5)
    #     url = 'http://www.ffiec.gov/census/report.aspx?year=2011&state=01&report=demographic&msa=11500'
        html = requests.get(browser.url).content
        print(browser.url)
        df_list = pd.read_html(html)
        data = df_list[-1]
        df1.append(data)
    #         print(df)
        time.sleep(1)
        # len(states_values)
    #     for x in states_values:
    #         one_state = browser.links.find_by_href(f'https://geology.com/county-map/{x.lower()}.shtml').click()
    #         html = browser.html
    #         soup = bs(html,'html.parser')
    #         image = soup.find('img')['src']
    #         images_urls.append({"state":x, "image":image})
    #         print()
    #         time.sleep(5)
    #         browser.back()
        browser.quit()
    return df1
df1 = scrape_list(first_list)
```

```python
dictionary_df1 = {}
count = 0
for i in df1:
    dictionary_df1[first_list[count]]= i.to_dict()
    count+= 1 
```
