```python
import pandas as pd
```

```python
twitch_df = pd.read_csv('Dota2_twitchData.csv')
twitch_df.columns
```

```python
twitch_df= twitch_df[['Month', 'Avg. viewers',
       'Peak_viewers', 'Avg. channels',
       'Peak_channels', 'Hours watched']]
twitch_df
```

```python
reddit_df = pd.read_csv('aggregated_dota2_reddit_data.csv')
reddit_df = reddit_df.sort_values(by=['utc date'])
reddit_df.columns
```

```python
reddit_df = reddit_df.rename(columns={'utc date':'Month'})
reddit_df
```

```python
files = ['DOTA2TOURN2016.csv','DOTA2TOURN2017.csv','DOTA2TOURN2018.csv','DOTA2TOURN2019.csv','DOTA2tourn2020.csv']
tournament_df = []
for onefile in files:
    temp_df = pd.read_csv(onefile, header = None)
    tournament_df.append(temp_df)
    
```

```python
cleaned_df = []
for onedf in tournament_df:
    testdf = onedf[0].str.split(" - ",expand=True)
    startdf = testdf[0].str.split("-",expand=True)
    startdf = startdf.rename(columns={0: 'start_year', 1: 'start_month',2: 'start_day'})
    enddf = testdf[1].str.split("-",expand=True)
    enddf = enddf.rename(columns={0: 'end_year', 1: 'end_month',2: 'end_day'})
    cleaned_df.append(pd.concat([onedf, startdf, enddf], axis=1))
# test_df = tournament_df[0][0].str.split(" - ",expand=True)
```

```python
# cleaned_df[0]
test2 = cleaned_df[0][['start_year','start_month']].agg('-'.join, axis=1)
```

```python
df = []
for one in cleaned_df:
    test2 = one[['start_year','start_month']].agg('-'.join, axis=1)
    df.append(test2)
    
```

```python
count = 0
for one in cleaned_df:
    one['end_date']=df[count]
    count+=1
```

```python
gb_df = pd.concat([cleaned_df[0],cleaned_df[1],cleaned_df[2],cleaned_df[3],cleaned_df[4]])
```

```python
gb_df.columns
```

```python
gb_df = gb_df[[             1,             2,    'end_date']]
gb_df = gb_df.rename(columns={1:'tournament', 2 : 'price'}).reset_index()
```

```python
gb_df = gb_df.drop(['index'], axis = 1)
```

```python
price_df = gb_df['price'].str.split("$",expand=True)
```

```python
gb_df = pd.concat([gb_df,price_df], axis = 1)
```

```python
# gb_df = gb_df.drop([0, 'price'], axis = 1)
gb_df = gb_df.rename(columns={1:'pricenum'})
```

```python
gb_df['price'] = gb_df['price'].apply(lambda x: x.replace('.00','')).apply(lambda x: x.replace(',',''))
# .astype('int32').replace(',','')
# gb_df = gb_df.drop(0,axis = 1)
gb_df
```

```python
gb_df_copy = gb_df.copy()
```

```python
for y in range(10,99):
    temp_replace = '.'+ str(y)
    gb_df_copy['price'] = gb_df_copy['price'].apply(lambda x: x.replace(temp_replace,''))
for y in range(1,9):
    temp_replace = '.0'+ str(y)
    gb_df_copy['price'] = gb_df_copy['price'].apply(lambda x: x.replace(temp_replace,''))
gb_df_copy
```

```python
gb_df_copy['price'] = gb_df_copy['price'].apply(lambda x: x.replace('$',''))
gb_df_copy
```

```python
gb_df_copy['price'] = gb_df_copy['price'].astype(int)
gb_df_copy
```

```python
gb_df_copy = gb_df_copy.drop([0,'pricenum'],axis = 1)
gb_df_copy
```

```python
final_gbdf = gb_df_copy.groupby(['end_date']).sum()
final_gbdf
```

```python
final_gbdf = final_gbdf.reset_index()
```

```python
dates=[]
from datetime import datetime
for x in final_gbdf['end_date']:
    dates.append(datetime.strptime(x, '%Y-%m'))
    
# final_gbdf['Month'] = datetime.strptime(final_gbdf['end_date'], '%Y-%m')
```

```python
final_gbdf['Month'] = dates
```

```python
final_gbdf = final_gbdf.drop('end_date', axis = 1)
```
