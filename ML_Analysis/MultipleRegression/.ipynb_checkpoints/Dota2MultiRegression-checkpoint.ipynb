```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Read the csv file into a pandas DataFrame
# Drop NA values, then display the first 5 records

dotaData = pd.read_csv('Dota2_final_df.csv')
dotaData.dropna(inplace=True)
dotaData.head()
```

```python
#With Peak Players
X = dotaData[["price", "Avg. viewers", "Peak_viewers", "Avg. channels", "Peak_channels", "Hours watched", "score_sum", "score_mean", "num comments_sum", "num comments_mean", "afinn_score_title_sum", "afinn_score_text_sum", "afinn_score_title_mean", "afinn_score_text_mean", "Peak Players"]]
y = dotaData["Avg. Players"].values.reshape(-1, 1)
print("Peak Players Included")
print("-----------------------")
print("-Data Shape-")
print(X.shape, y.shape)
print("")

# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Create the model using LinearRegression
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Fit the model to the training data and calculate the scores for the training and testing data
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)
 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")

# Plot the Residuals for the Training and Testing data
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Average Players Residual Plot")
```

# Key Takeaways
# ----------------------

## From the testing datasets, our model's accuracy improves if we include TIME and PEAK PLAYERS. 
### However, time is included intrinsically in some of the other data points in our dataset, and peak players is derived off the same metric as average players (time and number of players).

## Overall, it seems like there may be more to the picture than the data we gathered, which makes sense because the determining factors behind the average number of players during a given month is multi-faceted
### Some examples that our dataset may not consider:
#### - Was there major downtime for the game? Was there a major update to the game?
#### - Did a major event happen with a competitor?
#### - Were there any macro factors that could contribute to an increase or decline?
